```python
#While we check the dataset information, it is discovered that the Salary column is an oject
#instead of a feature. To clean this, we have to convert the salary column to type int and also perform 
#some conversion on the values of the column.

```

**Selecting the different salaries by month, year and hour**

```python
df_month = salaries.loc[salaries["Salary"].str.contains('/mo')]
df_year = salaries.loc[salaries["Salary"].str.contains('/yr')]
df_hour = salaries.loc[salaries["Salary"].str.contains('/hr')]
```

```python
df_month
```

**Removing unwanted characters from "Salary" column**

```python

    if df_month["Salary"].str.contains('/mo').any():
            df_month["Salary"] = df_month["Salary"].str.replace("/mo", ' ')
            df_month["Salary"] = df_month["Salary"].str.replace("₹", ' ')
            df_month["Salary"] = df_month["Salary"].str.replace("$", ' ')
            df_month["Salary"] = df_month["Salary"].str.replace(",", '')
            df_month["Salary"] = df_month["Salary"].str.replace("£", '')
            

    if df_year["Salary"].str.contains('/yr').any():
            df_year["Salary"] = df_year["Salary"].str.replace("/yr", ' ')
            df_year["Salary"] = df_year["Salary"].str.replace("₹", ' ')
            df_year["Salary"] = df_year["Salary"].str.replace("$", ' ')
            df_year["Salary"] = df_year["Salary"].str.replace(",", '')
            df_year["Salary"] =df_year["Salary"].str.replace("£", '')
            df_year["Salary"] =df_year["Salary"].str.replace("AFN", '')
    
    
    if df_hour["Salary"].str.contains('/hr').any():
            df_hour["Salary"] = df_hour["Salary"].str.replace("/hr", ' ')
            df_hour["Salary"] = df_hour["Salary"].str.replace("₹", ' ')
            df_hour["Salary"] = df_hour["Salary"].str.replace("$", ' ')
            df_hour["Salary"] = df_hour["Salary"].str.replace(",", '')
            df_hour["Salary"] = df_hour["Salary"].str.replace("£", '')
            df_hour["Salary"] = df_hour["Salary"].str.replace("AFN", '')
  
  
            
```

**Converting the datatype of the Salary columns to int from object**

```python
df_month["Salary"] = df_month['Salary'].apply(pd.to_numeric)
df_year["Salary"] = df_year['Salary'].apply(pd.to_numeric)
df_hour["Salary"] = df_hour['Salary'].apply(pd.to_numeric)
```

```python
df_month
```

**Conversion of Salaries per hour and monthly to their yearly equivalent**

```python
df_month["Salary"] = df_month["Salary"] *12
df_hour["Salary"] = df_hour["Salary"] * 8760
```

```python
df_month
```

**Appending the dataframes for the cleaned salary columns to csv files and merging them together into a new dataframe**

```python
df_year.to_csv("year.csv", index=False)
df_month.to_csv("month.csv", index=False)
df_hour.to_csv("hour.csv", index=False)
```

```python
df1 = pd.read_csv("year.csv")
df2 = pd.read_csv("month.csv")
df3 = pd.read_csv("hour.csv")
```

```python
df4 = pd.merge(df1, df2, how = "outer")
```

```python
df = pd.merge(df4, df3, how = "outer")
```

```python
df
```

**Checking the dataset information for missing values. Company name and Salaries Reported are discovered to have missing values**

```python
df.info()
```

**Filling the missing values in Salaries Reported with it's mean**

```python
df["Salaries Reported"] = df["Salaries Reported"].fillna(df["Salaries Reported"].mean())
```

**Dropping the rest rows which have missing values in Company name**

```python
df = df.dropna()
```
